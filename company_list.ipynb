# Scraping Data from Wiki

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text,'html')
```

```python
soup.find_all('table',class_='wikitable sortable')[1]
#pick table with index is 1, class is wikitable sortable 
```

```python
table = soup.find_all('table')[1]
# print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
# list titles of the table
```

```python
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
# create a dataframe
# df
```

```python
row_count = table.find_all('tr')
```

```python
for row in row_count[1:]: #skip the first empty row
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    # add each row
    
    length = len(df)
    df.loc[length] = individual_row_data # add each data cell for each row
```

```python
df
```

```python
df.to_csv(r'/data/company_list.csv', index = False)
# export the csv file into data folder
```
